```python
import os
from collections import Counter
import yaml

# Buka YAML untuk ambil nama kelas
with open('/content/Sampah-Organik-dan-Anorganik-3/data.yaml', 'r') as f:
    data_yaml = yaml.safe_load(f)
class_names = data_yaml['names']

# Ganti path ini ke direktori label kamu
label_dir = '/content/Sampah-Organik-dan-Anorganik-3/train/labels'

label_counter = Counter()

for label_file in os.listdir(label_dir):
    if label_file.endswith('.txt'):
        with open(os.path.join(label_dir, label_file)) as f:
            for line in f:
                class_id = int(line.strip().split()[0])
                label_counter[class_id] += 1

# Cetak hasilnya
print("Jumlah anotasi per kelas:")
for class_id, count in label_counter.items():
    class_name = class_names[class_id]
    print(f"{class_name} (ID: {class_id}): {count} anotasi")
```

```python
import os

label_dir = '/content/Sampah-Organik-dan-Anorganik-3/train/labels'
image_dir = '/content/Sampah-Organik-dan-Anorganik-3/train/images'

label_id_dicari = '0'  # Anorganik (sesuai data.yaml)

hasil = []

for file in os.listdir(label_dir):
    if file.endswith('.txt'):
        label_path = os.path.join(label_dir, file)
        with open(label_path, 'r') as f:
            lines = f.readlines()
            for line in lines:
                if line.strip().startswith(label_id_dicari):
                    # Ganti ekstensi .txt jadi .jpg atau .png sesuai dataset
                    image_name = file.replace('.txt', '.jpg')
                    image_path = os.path.join(image_dir, image_name)
                    hasil.append(image_path)
                    break  # cukup sekali per file

print(f"Ditemukan {len(hasil)} gambar dengan label ID {label_id_dicari} (Anorganik):")
for path in hasil[:10]:  # tampilkan 10 pertama
    print(path)
```
